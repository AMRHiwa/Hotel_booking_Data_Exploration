finding knowleg about missing value percentages each columns in DataFrame 

```python
# finding the mean of missing value in each column
missing_percentages = df.isnull().mean()
missing_percentages
```

finding index of columns that them missing value percentage is bigger than 90% 

```python
columns_to_delete = missing_percentages[missing_percentages > .9].index
columns_to_delete
```

deleting the columns that them missing values percentages is bigger than 0.9 from DataFrame

```python
df.drop(columns = columns_to_delete, inplace = True)
```

finding columns name of DataFrame that have same null number of missing value

```python
df.columns[df[df.adults.isna()].isna().sum() == df.adults.isna().sum()]
```

deleting samples that they have null value in 'adults' column

```python
df.dropna(subset = ['adults'],  inplace = True)
```

calculate mean/median of grouping dataframe by 'arrival_date_month' and 'arrival_date_week_number'

then filling null cell from 'arrival_date_week_number' with them mean/median

```python
mean = df.groupby('arrival_date_month')['arrival_date_week_number'].transform('mean')
df['arrival_date_week_number'].fillna(mean.astype('int64'), inplace = True)
```

```python
median = df.groupby('arrival_date_month')['arrival_date_week_number'].transform('median')
df['arrival_date_week_number'].fillna(median.astype('int64'), inplace = True)
```

finding mode of grouped dataframe by 'agent' column and 'country' column

```python
frequent_types = df.groupby('agent')['country'].transform(lambda x: x.mode().iloc[0])
df['country'].fillna(frequent_types, inplace = True)
```

calculate missing values of each columns from DataFrame again

```python
df.isna().sum().sum()
```

finding the outliers from dataframe by finding quantile 25%, 50%, 75% and calculating IQR

```python
# calculate quantile of 25% , 50% and 75%
Q1, Q2, Q3 = df.stays_in_week_nights.quantile([0.25, 0.5, 0.75]) 

# calculating IQR
IQR = Q3 - Q1

# create a new column names 'stays_in_week_nights_outlier' that is in below condition
df['stays_in_week_nights_outlier'] = df['stays_in_week_nights'].apply(lambda p: (p > Q3 + 1.5 * IQR) or (p < Q1 - 1.5 * IQR))

# showing the 10 random sample from dataframe with 'stays_in_week_nights' and 'stays_in_week_nights_outlier' columns
df[['stays_in_week_nights', 'stays_in_week_nights_outlier']].sample(10)
```

```python
# import plotly module for drawing box plot of outliers
import plotly.express as px 

# create figure of box plot by 'stays_in_week_nights' column 
fig = px.box(df, y='stays_in_week_nights') 

# adding marker color to our chart
fig.update_traces(marker_color='#FF851B')

# adding the width, height and title features to our chart
fig.update_layout(autosize=False, width=600, height=500, title='Detected Outliers') 

# save boxplot
plt.savefig("outliers_boxplot.jpg")

# showing the chart
fig.show()
```

create a new 'kids' column by sum of 'children' and 'babies' columns. 

and delete the 'children' and 'babies' column from dataframe.

```python
df['kids'] = df['children'] + df['babies']
df.drop(columns = ['children', 'babies'], inplace = True)
```

create a new column names 'family' with 1 and 0 value that equals with 1 if 'kids' feature of sample is bigger than 0, else equals with 0.

```python
df['family'] = np.where(df['kids'] > 0 , 1 , 0)
```

showing the countplot for number of any type of family for each months

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the percentage of each 'arrival_date_month' within the 'family' == 0 group
family_0 = df[df['family'] == 0]
family_0_counts = family_0['arrival_date_month'].value_counts(normalize=True) * 100
family_0_counts = family_0_counts.reset_index()
family_0_counts.columns = ['arrival_date_month', 'percent']

# Calculate the percentage of each 'arrival_date_month' within the 'family' == 1 group
family_1 = df[df['family'] == 1]
family_1_counts = family_1['arrival_date_month'].value_counts(normalize=True) * 100
family_1_counts = family_1_counts.reset_index()
family_1_counts.columns = ['arrival_date_month', 'percent']

# Create a barplot for all samples where 'family' feature is equal to 0
sns.barplot(data=family_0_counts, x='arrival_date_month', y='percent')
plt.title('Percentage of Arrival Date Month for Family = 0')
plt.xticks(rotation=90)

# save countplot
plt.savefig("Percentage_of_Arrival_Date_Month_for_Family_0.png")

plt.show()

# Create a barplot for all samples where 'family' feature is equal to 1
sns.barplot(data=family_1_counts, x='arrival_date_month', y='percent')
plt.title('Percentage of Arrival Date Month for Family = 1')
plt.xticks(rotation=90)

# save countplot
plt.savefig("Percentage_of_Arrival_Date_Month_for_Family_1.png")

plt.show()
```

Due to the large dispersion of the lead_time column, we divide this column into meaningful categories.

```python
# Define the bins for categorizing 'lead_time' based on specific ranges
bins = [df['lead_time'].min(), 7, 14, 30, 90, 120, 365, df['lead_time'].max()]

# Categorize 'lead_time' into bins and create a new column 'lead_time_bins'
df['lead_time_bins'] = pd.cut(df['lead_time'], bins)

# Define the labels for the bins
maps = ['less than a week', 'less than 2 weeks', 
        'less than a month', 'less than 3 months', 'less than 6 months',
        'less than a year', 'more than a year']

# Create a dictionary to map the bins to their corresponding labels
map_dict = {bin: label for bin, label in zip(df['lead_time_bins'].sort_values().unique(), maps)}

# Apply the mapping to the 'lead_time_bins' column to replace bin ranges with labels
df['lead_time_bins'] = df['lead_time_bins'].map(map_dict)
```

drawing the barplot for lead_time_bins

```python
# Group the data by 'lead_time_bins' and calculate the mean of 'is_canceled' for each bin
grouped = df.groupby('lead_time_bins', observed=True)['is_canceled'].mean()

# Create a bar plot to visualize the mean 'is_canceled' values for each 'lead_time_bins'
grouped.plot(kind='bar');

# save barplot
plt.savefig("lead_time_bin.png")

```
